# Forecasting

## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Reading Data

```python
df = pd.read_excel("./Data.xlsx")
```

```python
df.head()
```

## Data Cleaning

```python
df.drop([df.columns[1]],axis=1, inplace=True)
df.drop(0,inplace=True)
df["Date"] = pd.to_datetime(df["Date"])
df.set_index(["Date"], inplace=True)
df.head()
```

```python
df.shape
```

## Data Exploration

```python
_, ax = plt.subplots(2,2,figsize=(10, 10))
ax[0][0].plot(df["Open"], label="Open")
ax[0][0].set_title("Stock Open Time Series")
ax[0][1].plot(df["High"], label="High")
ax[0][1].set_title("Stock High Time Series")
ax[1][0].plot(df["Low"],  label="Low")
ax[1][0].set_title("Stock Low Time Series")
ax[1][1].plot(df["Close"],label="Close")
ax[1][1].set_title("Stock Close Time Series")
for i in ax:
    for j in i:
        j.legend()
        for tick in j.get_xticklabels():
            tick.set_rotation(45)
plt.tight_layout()
plt.show()
```
