### Checking Delta for Each Day

````python
df["delta"] = df.apply(lambda x: x["Close"] - x["Open"],axis=1)
````

## Data Exploration

### Creating Predefined Function

````python
def exploration_function(values, col):
    # Series Plotting
    _, ax = plt.subplots(1,1,figsize=(10, 10))
    ax.plot(values,label=col)
    ax.set_title(f"Stock {col} Time Series")
    for tick in ax.get_xticklabels():
        tick.set_rotation(90)
    plt.tight_layout()
    plt.show()
    
    add = seasonal_decompose(values,model="Additive", period=1)
    plt.rcParams.update({'figure.figsize': (10,10)})
    add.plot()
    plt.show()
    try:
        mul = seasonal_decompose(values,model="multiplicative", period=1)
        plt.rcParams.update({'figure.figsize': (10,10)})
        mul.plot()
    except Exception as e:
        print("For multiplicative series negative values are not allowed.")
    
    _, ax = plt.subplots(1,2, figsize=(20,10))
    plot_acf(values, lags=50, ax=ax[0])
    plot_pacf(values, lags=50, ax=ax[1])
    plt.show()
    
    result = adfuller(values, autolag="AIC")
    print(f'ADF Statistic: {result[0]}')
    print(f'n_lags: {result[1]}')
    print(f'p-value: {result[1]}')
    for key, value in result[4].items():
        print('Critial Values:')
        print(f'   {key}, {value}')        
````

### Checking the Series combined

````python
_, ax = plt.subplots(1,1,figsize=(10, 10))
ax.plot(df["Open"], label="Open")
ax.plot(df["High"], label="High")
ax.plot(df["Low"],  label="Low")
ax.plot(df["Close"],label="Close")
for tick in ax.get_xticklabels():
    tick.set_rotation(90)
ax.legend()
ax.grid()
plt.show()
````

**Observations**:
- From the above graphs give Time Series Data does not look stationary

### Checking Individual Series

- **Null Hypothesis ($H_{o}$)** : The Series contains a **Unit Root** and hence the series is non **Stationary**
- **Alternate Hypothesis ($H_{1}$)** : The Series does not contain a **Unit Root** and hence the series is **Stationary**

#### Open ( Original Series )

````python
exploration_function(df["Open"], col="Open")
````

**Observation**: Since **p-value** is greater than significance level of 0.05 and value of the statistic is greater than critical values then we can safely say that, **We fail to reject the null hypothesis and hence the series is Non-Stationary**

#### Close ( Original Series ) 

````python
exploration_function(df["Close"], col="Close")
````

**Observation**: Since **p-value** is greater than significance level of 0.05 and value of the statistic is greater than critical values then we can safely say that, **We fail to reject the null hypothesis and hence the series is Non-Stationary**

#### Open ( With diff =1 ) 

````python
exploration_function(df["Open"].diff().dropna(), col="Open")
````

**Observation**: Since **p-value** is lower than the significance level of 0.05 and value of the statistic is smaller than critical values then we can safely say that, **We accept the null hypothesis and hence the series is Stationary**

#### Close ( With diff =1)

````python
exploration_function(df["Close"].diff().dropna(), col="Close")
````

**Observation**: Since **p-value** is lower than the significance level of 0.05 and value of the statistic is smaller than critical values then we can safely say that, **We accept the null hypothesis and hence the series is Stationary**

#### Delta

````python
exploration_function(df["delta"], col="delta")
````

**Observation**: Since **p-value** is lower than the significance level of 0.05 and value of the statistic is smaller than critical values then we can safely say that, **We accept the null hypothesis and hence the series is Stationary**

**But**

Let's take this with a ```grain of salt```  because as we can see in the **Auto Correlation** and **Partial Auto Correlation** plots that there is not much correlation with previous value hence we will consider the delta series as ```white noise```.
