## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from dateutil.parser import parse
```

## Reading Data

```python
df = pd.read_excel("./Data.xlsx")
```

```python
df.head()
```

## Data Cleaning

```python
df.drop([df.columns[1]],axis=1, inplace=True)
df.drop(0,inplace=True)
df["Date"] = pd.to_datetime(df["Date"])
df.set_index(["Date"], inplace=True)
df.dropna(inplace=True)
df.head()
```

```python
df.shape
```

```python
df[df.isnull().any(1)]
```

## Data Exploration

### Checking Individual Series

```python
_, ax = plt.subplots(2,2,figsize=(10, 10))
ax[0][0].plot(df["Open"], label="Open")
ax[0][0].set_title("Stock Open Time Series")
ax[0][1].plot(df["High"], label="High")
ax[0][1].set_title("Stock High Time Series")
ax[1][0].plot(df["Low"],  label="Low")
ax[1][0].set_title("Stock Low Time Series")
ax[1][1].plot(df["Close"],label="Close")
ax[1][1].set_title("Stock Close Time Series")
for i in ax:
    for j in i:
        j.legend()
        for tick in j.get_xticklabels():
            tick.set_rotation(90)
plt.tight_layout()
plt.show()
```

### Checking the Series combined

```python
_, ax = plt.subplots(1,1,figsize=(10, 10))
ax.plot(df["Open"], label="Open")
ax.plot(df["High"], label="High")
ax.plot(df["Low"],  label="Low")
ax.plot(df["Close"],label="Close")
for tick in ax.get_xticklabels():
    tick.set_rotation(90)
ax.legend()
ax.grid()
plt.show()
```

**Observations**:
- From the above graphs give Time Series Data does not look stationary

### Checking Delta for Each Day

$\delta = x_{closing} - x_{opening}$

```python
df["delta"] = df.apply(lambda x: x["Close"] - x["Open"],axis=1)
```

```python
df.head()
```

```python
_, ax = plt.subplots(1,1,figsize=(10, 10))
ax.plot(df["delta"], label="delta")
for tick in ax.get_xticklabels():
    tick.set_rotation(90)
ax.legend()
ax.grid()
plt.show()
```

**Observations**:
- The delta time series looks somewhat like a stationary series

```python
df["month"] = df.index.month_name()
```

```python
df["year"] = df.index.year
```

```python
df.head()
```

```python
temp = df.groupby(["month", "year"])["delta"].sum().reset_index()
month = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
temp['month'] = pd.Categorical(temp['month'], categories=month, ordered=True)
# temp.sort_values(["year", "month"], inplace=True)
temp.head()
```

```python
_, ax = plt.subplots(1,1,figsize=(10, 10))
sns.lineplot("month","delta", hue="year",  data=temp, ax=ax)
for tick in ax.get_xticklabels():
    tick.set_rotation(90)
ax.legend()
ax.grid()
plt.show()
```

```python
temp = df.groupby(["month", "year"])["Open"].sum().reset_index()
month = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
temp['month'] = pd.Categorical(temp['month'], categories=month, ordered=True)
temp.head()
```

```python
_, ax = plt.subplots(1,1,figsize=(10, 10))
sns.lineplot("month","Open", hue="year",  data=temp, ax=ax)
for tick in ax.get_xticklabels():
    tick.set_rotation(90)
ax.legend()
ax.grid()
plt.show()
```

**Observation**:
- No seasonality pattern visible

```python
temp = df.groupby(["month", "year"])["Close"].sum().reset_index()
month = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
temp['month'] = pd.Categorical(temp['month'], categories=month, ordered=True)
temp.head()
```

```python
_, ax = plt.subplots(1,1,figsize=(10, 10))
sns.lineplot("month","Close", hue="year",  data=temp, ax=ax)
for tick in ax.get_xticklabels():
    tick.set_rotation(90)
ax.legend()
ax.grid()
plt.show()
```

### Decomposing a Time Series

```python
add = seasonal_decompose(df["Open"],model="Additive", period=1)
plt.rcParams.update({'figure.figsize': (10,10)})
add.plot()
```

```python
mul = seasonal_decompose(df["Open"],model="multiplicative", period=1)
plt.rcParams.update({'figure.figsize': (10,10)})
mul.plot()
```

```python
add = seasonal_decompose(df["delta"],model="Additive", period=1)
plt.rcParams.update({'figure.figsize': (10,10)})
add.plot()
```
